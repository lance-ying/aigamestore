```python
import os
import json
import glob
from pathlib import Path
import re
```

```python

def extract_and_save_concepts():
    # Define base path for input and output
    base_path = '.'
    output_path = '../finalized_concepts'
    
    # Create output directory if it doesn't exist
    os.makedirs(output_path, exist_ok=True)
    
    # Get all folders in the base path
    folders = os.listdir(base_path)
    
    # Counter for new file naming
    counter = 0
    
    # Process each folder
    for folder in folders:
        folder_path = os.path.join(base_path, folder)
        if not os.path.exists(folder_path):
            print(f"Folder {folder_path} does not exist, skipping.")
            continue
        
        # Get all JSON files in the folder
        json_files = glob.glob(os.path.join(folder_path, "*.json"))
        
        for json_file in json_files:
            try:
                # Read the JSON file
                with open(json_file, 'r', encoding='utf-8') as f:
                    data = json.load(f)
                
                # Extract the concept field
                if 'concept' in data:
                    # Create new data with only the concept field and add model info
                    new_data = {
                        "concept": data["concept"],
                    }                                        
                    # Create new filename
                    new_filename = f"game_{counter:04d}.json"
                    new_filepath = os.path.join(output_path, new_filename)
                    
                    # Save the new JSON file
                    with open(new_filepath, 'w', encoding='utf-8') as f:
                        json.dump(new_data, f, indent=2)
                    
                    counter += 1
                    
            except Exception as e:
                print(f"Error processing {json_file}: {e}")
    
    print(f"Processed {counter} game concepts and saved to {output_path}")

# Run the extraction function
extract_and_save_concepts()
```

```python
def read_and_print_concepts(concepts_dir="../finalized_concepts"):
    """
    Read through all the concept files in the specified directory and print them
    with the format: filename: concept
    
    Args:
        concepts_dir (str): Directory containing the concept JSON files
    """
    # Check if directory exists
    if not os.path.exists(concepts_dir):
        print(f"Directory {concepts_dir} does not exist.")
        return
    
    # Get all JSON files in the directory
    json_files = glob.glob(os.path.join(concepts_dir, "*.json"))
    
    if not json_files:
        print(f"No JSON files found in {concepts_dir}.")
        return
    
    # Sort files by name to ensure consistent ordering
    json_files.sort()
    
    print(f"Found {len(json_files)} concept files in {concepts_dir}:\n")
    
    # Read and print each concept
    for json_file in json_files:
        try:
            # Get the filename without the path
            filename = os.path.basename(json_file)
            
            # Read the JSON file
            with open(json_file, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            # Extract and print the concept
            if 'concept' in data:
                print(f"{filename.split('.')[0]}: {data['concept']}")
            else:
                print(f"{filename.split('.')[0]}: [No concept field found]")
                
        except Exception as e:
            print(f"Error reading {json_file}: {e}")
    
    print(f"\nFinished printing {len(json_files)} concepts.")

# Run the function to read and print all concepts
read_and_print_concepts()
```
